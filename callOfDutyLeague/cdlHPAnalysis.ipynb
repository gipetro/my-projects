Here we have a dataset that records the scores of matches in the Call of Duty League, specifically in the gamemode "Hardpoint". I want to explore with this where teams are performing the best, the worst, how they score their points, and other nuances of this fast paced game.

To start, the data will be transformed so that the points are shown by the amount earned each hill, rather than the total score at that point in the game.

```python
timePerHill = bo7raw.copy()

timePerHill['Hill 13 A'] -= timePerHill['Hill 12 A']
timePerHill['Hill 13 B'] -= timePerHill['Hill 12 B']

timePerHill['Hill 12 A'] -= timePerHill['Hill 11 A']
timePerHill['Hill 12 B'] -= timePerHill['Hill 11 B']

timePerHill['Hill 11 A'] -= timePerHill['Hill 10 A']
timePerHill['Hill 11 B'] -= timePerHill['Hill 10 B']

timePerHill['Hill 10 A'] -= timePerHill['Hill 9 A']
timePerHill['Hill 10 B'] -= timePerHill['Hill 9 B']

timePerHill['Hill 9 A'] -= timePerHill['Hill 8 A']
timePerHill['Hill 9 B'] -= timePerHill['Hill 8 B']

timePerHill['Hill 8 A'] -= timePerHill['Hill 7 A']
timePerHill['Hill 8 B'] -= timePerHill['Hill 7 B']

timePerHill['Hill 7 A'] -= timePerHill['Hill 6 A']
timePerHill['Hill 7 B'] -= timePerHill['Hill 6 B']

timePerHill['Hill 6 A'] -= timePerHill['Hill 5 A']
timePerHill['Hill 6 B'] -= timePerHill['Hill 5 B']

timePerHill['Hill 5 A'] -= timePerHill['Hill 4 A']
timePerHill['Hill 5 B'] -= timePerHill['Hill 4 B']

timePerHill['Hill 4 A'] -= timePerHill['Hill 3 A']
timePerHill['Hill 4 B'] -= timePerHill['Hill 3 B']

timePerHill['Hill 3 A'] -= timePerHill['Hill 2 A']
timePerHill['Hill 3 B'] -= timePerHill['Hill 2 B']

timePerHill['Hill 2 A'] -= timePerHill['Hill 1 A']
timePerHill['Hill 2 B'] -= timePerHill['Hill 1 B']

timePerHill
```

```python
bo7raw3 = pd.read_csv('cdlHillCountsBO7v3.csv')
timePerHill3 = bo7raw3.copy()

timePerHill3['Hill 13 A'] -= timePerHill3['Hill 12 A']
timePerHill3['Hill 13 B'] -= timePerHill3['Hill 12 B']
timePerHill3['Hill 12 A'] -= timePerHill3['Hill 11 A']
timePerHill3['Hill 12 B'] -= timePerHill3['Hill 11 B']
timePerHill3['Hill 11 A'] -= timePerHill3['Hill 10 A']
timePerHill3['Hill 11 B'] -= timePerHill3['Hill 10 B']
timePerHill3['Hill 10 A'] -= timePerHill3['Hill 9 A']
timePerHill3['Hill 10 B'] -= timePerHill3['Hill 9 B']
timePerHill3['Hill 9 A'] -= timePerHill3['Hill 8 A']
timePerHill3['Hill 9 B'] -= timePerHill3['Hill 8 B']
timePerHill3['Hill 8 A'] -= timePerHill3['Hill 7 A']
timePerHill3['Hill 8 B'] -= timePerHill3['Hill 7 B']
timePerHill3['Hill 7 A'] -= timePerHill3['Hill 6 A']
timePerHill3['Hill 7 B'] -= timePerHill3['Hill 6 B']
timePerHill3['Hill 6 A'] -= timePerHill3['Hill 5 A']
timePerHill3['Hill 6 B'] -= timePerHill3['Hill 5 B']
timePerHill3['Hill 5 A'] -= timePerHill3['Hill 4 A']
timePerHill3['Hill 5 B'] -= timePerHill3['Hill 4 B']
timePerHill3['Hill 4 A'] -= timePerHill3['Hill 3 A']
timePerHill3['Hill 4 B'] -= timePerHill3['Hill 3 B']
timePerHill3['Hill 3 A'] -= timePerHill3['Hill 2 A']
timePerHill3['Hill 3 B'] -= timePerHill3['Hill 2 B']
timePerHill3['Hill 2 A'] -= timePerHill3['Hill 1 A']
timePerHill3['Hill 2 B'] -= timePerHill3['Hill 1 B']

timePerHill3[timePerHill3.columns[5:31]].min()
```

Looks good! Now, based off of what we just looked at, the first point I want to explore is the amount of shutouts occuring for each hill. From that we can get a get an idea of what points tend to be the easiest to defend, and what points have the most back and forth fighting taking place.

```python
shutouts = (timePerHill3[timePerHill3.columns[5:31]] == 0).sum()
shutouts
```

Ok, so here we get an idea of how many shutouts are hapenning in general, but we can't learn much from this. What would be better is if we grouped them by map. Then, knowing each map only has 4 or 5 actual hill locations, we can sum each unique hill with the others that ocurred at the same location.

```python
mapLook = timePerHill3[timePerHill3.columns[4:31]]
mapLook.iloc[:, 1:27] = mapLook.iloc[:, 1:27].apply(lambda x: x==0)
mapLook = mapLook.groupby('Map')[mapLook.columns[1:27]].sum().reset_index()

fourHillMaps = ['Colossus', 'Den']
fiveHillMaps = ['Scar', 'Exposure', 'Blackheart']

hps4 = mapLook[mapLook['Map'].isin(fourHillMaps)].copy()
hps5 = mapLook[mapLook['Map'].isin(fiveHillMaps)].copy()

hps4['P1'] = hps4['Hill 1 A']+hps4['Hill 1 B'] + hps4['Hill 5 A']+hps4['Hill 5 B'] + hps4['Hill 9 A']+hps4['Hill 9 B'] + hps4['Hill 13 A']+hps4['Hill 13 B']
hps4['P2'] = hps4['Hill 2 A']+hps4['Hill 2 B'] + hps4['Hill 6 A']+hps4['Hill 6 B'] + hps4['Hill 10 A']+hps4['Hill 10 B']
hps4['P3'] = hps4['Hill 3 A']+hps4['Hill 3 B'] + hps4['Hill 7 A']+hps4['Hill 7 B'] + hps4['Hill 11 A']+hps4['Hill 11 B']
hps4['P4'] = hps4['Hill 4 A']+hps4['Hill 4 B'] + hps4['Hill 8 A']+hps4['Hill 8 B'] + hps4['Hill 12 A']+hps4['Hill 12 B']

hps5['P1'] = hps5['Hill 1 A']+hps5['Hill 1 B'] + hps5['Hill 6 A']+hps5['Hill 6 B'] + hps5['Hill 11 A']+hps5['Hill 11 B']
hps5['P2'] = hps5['Hill 2 A']+hps5['Hill 2 B'] + hps5['Hill 7 A']+hps5['Hill 7 B'] + hps5['Hill 12 A']+hps5['Hill 12 B']
hps5['P3'] = hps5['Hill 3 A']+hps5['Hill 3 B'] + hps5['Hill 8 A']+hps5['Hill 8 B'] + hps5['Hill 13 A']+hps5['Hill 13 B']
hps5['P4'] = hps5['Hill 4 A']+hps5['Hill 4 B'] + hps5['Hill 9 A']+hps5['Hill 9 B'] 
hps5['P5'] = hps5['Hill 5 A']+hps5['Hill 5 B'] + hps5['Hill 10 A']+hps5['Hill 10 B']

hps4 = hps4[['Map', 'P1', 'P2', 'P3', 'P4']]
hps5 = hps5[['Map', 'P1', 'P2', 'P3', 'P4', 'P5']]
shutoutsByMap = pd.concat([hps4, hps5])
shutoutsByMap
```

There it is! Now we can get an idea of which hills are the "mixiest", like P1 Colossus, or P4 Scar. The biggest "money hills" appear to be P2s on Colossus, Scar, and Blackheart.

Though some of these findings aren't too surprising for keen observers of the CDL, we can use the same knowledge to question some of our findings as well. For instance, the first hill, "P1", is often set up to be an even battle for each team as it is the first of the game. This leads to scenarios where some teams avoid scoring points in favor of attempting to gain good positioning. This could be boosting the amount of times where shutouts occur.

It is also known that the map "Blackheart" was not able to be played for the beginning portion of the season due to bugs. This likely means that the "money hills" on this map are even more important given there was less recorded matches played there in comparison to others.
